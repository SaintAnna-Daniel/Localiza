```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'Y:\\Car Rental\\Localiza\\basket\\basket 28-03-25.xlsx'

df = pd.read_excel(file_path)
```

```python


plt.figure(figsize=(10, 6))
total = len(df['price'])
plt.hist(df['price'], bins = 20, weights = np.ones(len(df['price'])) * 100/total)
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Percentage')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python

counts, bin_edges = np.histogram(df['price'], bins=20)

# Calculate percentages
total = len(df['price'])
percentages = (counts / total) * 100

# Display bin information
bin_info = pd.DataFrame({
    'bin_start': bin_edges[:-1],
    'bin_end': bin_edges[1:],
    'count': counts,
    'percentage': percentages
})

print(bin_info)
```

```python
counts, bin_edges = np.histogram(df['modelYear'], bins=6)

# Calculate percentages
total = len(df['modelYear'])
percentages = (counts / total) * 100

# Display bin information
bin_info = pd.DataFrame({
    'bin_start': bin_edges[:-1],
    'bin_end': bin_edges[1:],
    'count': counts,
    'percentage': percentages
})

print(bin_info)
```
